###Task 1
Сгенерировать 100 псевдослучайных чисел на отрезке [0; 4] (встроенной функцией
генерации псевдослучайных чисел). Проверить гипотезу о равномерном распределении на [0; 4].

```python
import numpy

random_float_array = numpy.random.uniform(0, 4, size=100)
random_float_array
```

H0 = {"датчик псевдослучайных чисел на отрезке [0;4] имеет равномерное распределение"}

интервалы: [0; 1), [1; 2), [2; 3), [3; 4], r =4.

```python
count_for_delta = {'delta1': len([item for item in random_float_array if item >=0 and item < 1]), 'delta2': len([item for item in random_float_array if item >=1 and item < 2]), 'delta3': len([item for item in random_float_array if item >=2 and item < 3]), 'delta4': len([item for item in random_float_array if item >=3 and item <= 4])}
```

```python
pi0 = {'delta1': 0.25, 'delta2': 0.25, 'delta3': 0.25, 'delta4': 0.25}
```

```python
calculate = {'i':[1, 2, 3, 4], 'ni': list(count_for_delta.values()), 'pi0': list(pi0.values())}
```

```python
import pandas

df = pandas.DataFrame.from_dict(calculate)
```

```python
df['npi0'] = df['pi0']*100
df['ni-npi0'] = df['ni'] - df['npi0']
df['pearson'] = df['ni-npi0']**2 / df['npi0']
df
```

```python
summa_pearsona = sum(df['pearson'])
summa_pearsona
```

```python
x_krit = 7.8
```

```python
if summa_pearsona > x_krit:
  print("Нулевая гипотеза отвергается")
else:
  print("Нет оснований отвергнуть нулевую гипотезу")
```

###Task 2
Сгенерировать 400 псевдослучайных чисел на отрезке [0; 4] (встроенной функцией генерации псевдослучайных чисел), каждое возвести в квадрат. Проверить
гипотезу о равномерном распределении этих квадратов на [0; 16].


```python
random_float_array2 = numpy.random.uniform(0, 4, size=400)
random_float_array2 = [x**2 for x in random_float_array2]
random_float_array2
```

интервалы: [0; 5), [5; 10), [10; 13), [13; 16], r =4.

```python
count_for_delta = {'delta1': len([item for item in random_float_array2 if item >=0 and item < 5]), 
    'delta2': len([item for item in random_float_array2 if item >=5 and item < 10]),
    'delta3': len([item for item in random_float_array2 if item >=10 and item < 13]), 
    'delta4': len([item for item in random_float_array2 if item >=13 and item <= 16])}
pi0 = {'delta1': 5/16, 'delta2': 5/16, 'delta3': 3/16, 'delta4': 3/16}
calculate = {'ni': list(count_for_delta.values()), 'pi0': list(pi0.values())}
df2 = pandas.DataFrame.from_dict(calculate)
df2
```

```python
df2['npi0'] = df2['pi0']*400
df2['ni-npi0'] = df2['ni'] - df2['npi0']
df2['pearson'] = df2['ni-npi0']**2 / df2['npi0']
df2
```

```python
summa_pearsona2 = sum(df2['pearson'])
x_krit2 = 7.8
if summa_pearsona2 > x_krit2:
  print("Нулевая гипотеза отвергается")
else:
  print("Нет оснований отвергнуть нулевую гипотезу")
```
