###Task1
По предложенной таблице* (выборке) проверить гипотезу о равномерном распределении на отрезке [0; c] (для неизвестного параметра c).

```python
import xlrd
numbers = []
rb = xlrd.open_workbook('/content/drive/MyDrive/Colab Notebooks/terver/Практика 5.2 и 5.3 - данные для анализа.xls', formatting_info=True)
sheet = rb.sheet_by_index(0)
for rownum in range(sheet.nrows):
  row = sheet.row_values(rownum)
  for c_el in row:
    numbers.append(c_el)
numbers
```

```python
c = max(numbers)
c
```

интервалы [0;2), [2; 4), [4;6), [6;c)

```python
import pandas
count_for_delta = {'delta1': len([item for item in numbers if item >=0 and item < 2]), 
    'delta2': len([item for item in numbers if item >=2 and item < 4]),
    'delta3': len([item for item in numbers if item >=4 and item < 6]), 
    'delta4': len([item for item in numbers if item >=6 and item <= c])}
pi0 = {'delta1': 2/c, 'delta2': 2/c, 'delta3': 2/c, 'delta4': (c-6)/c}
calculate = {'ni': list(count_for_delta.values()), 'pi0': list(pi0.values())}
df1 = pandas.DataFrame.from_dict(calculate)
df1
```

```python
df1['npi0'] = df1['pi0']*len(numbers)
df1['ni-npi0'] = df1['ni'] - df1['npi0']
df1['pearson'] = df1['ni-npi0']**2 / df1['npi0']
df1
```

r = 4, k = 1
X_krit = x_krit(r-1-k, alfa)=x_krit(2, alfa)

```python
summa_pearsona1 = sum(df1['pearson'])
x_krit1 = 5.99146 
if summa_pearsona1 > x_krit1:
  print("Нулевая гипотеза отвергается")
else:
  print("Нет оснований отвергнуть нулевую гипотезу")
```
